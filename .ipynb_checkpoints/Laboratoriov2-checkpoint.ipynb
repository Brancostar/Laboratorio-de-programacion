### 1.3.1 Transformaci칩n de notas a "Nota Chilena" [0.5 punto]

En la ultima reunion con Don Caguayo este les comenta es complejo para el y su equipo entender la graduaci칩n de notas de 0 a 100. Por ende, le comenta que desde hace harto tiempo tienen la intenci칩n de *trnasformar* todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci칩n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci칩n entregada, cree un funci칩n que le permita pasar a la escala Chilena las notas que se escapan del rango [1,7]. 

(No ocupe ning칰n tipo de ciclo).

**To-Do**

- [ ] Tranforme los valores a la escala chilena.
- [ ] Asigne los valores transformados al DataFrame.

````python
def escala_chilena(x):
    if (x>=1 or x<=7):
        return x*0.06+1

df_nuevo['math score']=df_nuevo['math score'].apply(escala_chilena)
df_nuevo["reading score"]=df_nuevo["reading score"].apply(escala_chilena)
df_nuevo["writing score"]=df_nuevo["writing score"].apply(escala_chilena)

df_nuevo.head(50)
````

### 1.3.2 Obtenci칩n de promedio de alumnos [0.5 puntos]

Con todas las notas adjuntadas, la prestigiosa *Universidad De la Cacha침a* necesita obtener el promedio general de cada uno de los alumnos (Grade Point Average - GPA). Para esto se le solicita que calcule el promedio con cada una de las asignaturas y el promedio se anexe como una nueva columna al final del DataFrame original.

**To-Do:**
- [ ] Obtenga el promedio de cada uno de los alumnos.
- [ ] Anexe esta informaci칩n en el DataFrame original.


````python
prom_grades=np.around((df_nuevo['math score']+df_nuevo['reading score']+df_nuevo['writing score'])/3,1)
prom_grades.head(5)

df_new=pd.concat([df_nuevo,prom_grades],axis=1).copy()
df_new.columns=[                      'names',                      'gender',
                    'race/ethnicity', 'parental level of education',
                             'lunch',     'test preparation course',
                        'math score',               'reading score',
                    'writing score',                             'mean']
df_new.head(50)
````

### 1.3.3 Reprobados [1 punto]

Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena.

Cree una funci칩n que permita obtener dos Dataframes, el primero debe poseer solamente los alumnos que han reprobado en la prestigiosa universidad y el segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.

**To-Do:**

- [ ] Obtenga una m치scara DataFrame con los alumnos que reprobaron el a침o.
- [ ] Reemplaze todos los GPA reprobados con una **R**.

````python
def reprobados(df):
    a=df["mean"]<4.0
    r=df['names'][a]
    df.loc[a,'mean']='R'
    return r,df
````

### 1.3.4 Informaci칩n por Categor칤as [1 punto]

Motivado por saber m치s sobre los datos, el Rector les solicita que genere una funci칩n que le permita comparar los promedios de diferentes **atributos categ칩ricos** del DataFrame. Para esto su funci칩n deber치 tener como entrada el DataFrame con todos los promedios (sin agregar las R's) y calcular el promedio **agrupando por etnias y el nivel educacional de los padres**.

**To-Do:**

- [ ] Hacer una funci칩n que nos permita obtener la comparaci칩n de promedios para dos atributos categ칩ricos.
- [ ] Cada uno de los grupos expuestos deben exponer el porcentaje de alumnos que representan en el colegio.
- [ ] Comente lo que observa de los resultados.
- [ ] Visualice los datos generados con [`px.sunburst`](https://plotly.com/python/sunburst-charts/ ).


````python
def category(df):
    df2=np.around(df.groupby(['race/ethnicity', 'parental level of education']).mean(),2).copy()
    r=(df.groupby(['race/ethnicity', 'parental level of education']).count()*100/len(df))['names']
    r=pd.concat([df2,r],axis=1)
    r=r.rename({'names':'Porcentaje'},axis=1)
    return r
category(df_new)
````

Al observar la tabla es posible darse cuenta que aquellos estudiantes que poseen padres con un nivel de escuela de bachiller, poseen mejores notas en comparacion al resto, en el mismo sentido a medida que los grupos de etnias disminuyen (lexicograficamente) los alumnos poseen mejor calificacion que el resto.
